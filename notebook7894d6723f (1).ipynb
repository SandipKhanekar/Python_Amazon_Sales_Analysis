```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('/kaggle/input/amazon-sale-report-100/Amazon Sale Report.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
# Display the first few rows of the DataFrame
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#drop blank columns
df.drop(['New','PendingS'], axis =1, inplace = True)
```

```python
df.info()
```

```python
#checking null values
pd.isnull(df)
```

```python
#sum will give total values of null values
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
#sum will give total values of null values
pd.isnull(df).sum()
```

```python
df.columns
```

```python
#Change Data type of column
df['ship-postal-code'] = df['ship-postal-code'].astype('int')

#check Datatype
df['ship-postal-code'].dtype
```

```python
#change date Datatype
df['Date'] = pd.to_datetime(df['Date'])

#check Datatype
df['Date'].dtype
```

```python
#rename Columns
df.rename(columns ={'Qty':'Quantity'})
```

```python
#Describe() method return descritpion of the data in the data frame (i.e count, mean,std,mi..etc)
df.describe()

# describe objects
df.describe(include = 'object')
```

```python
#Describe specific columns
df[['Qty', 'Amount']].describe()
```

```python
#Data Analysis

df.columns
```

```python
ax= sns.countplot(x='Size',data=df)

for bars in ax.containers:
    ax.bar_label(bars)

# from below Graph you can see that most of the people buys M-size
```

```python
#Group By 
S_Qty= df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty', ascending = True)

sns.barplot(x='Size', y='Qty', data= S_Qty)
```

```python
#Courier Status
plt.figure (figsize=(10,5))
ax= sns.countplot(data=df, x='Courier Status',hue ='Status')
plt.show()

#Note: from below graph the majority of the orders are shipped through the courier
```

```python
#Histogram
df['Size'].hist()
```

```python
##category

df['Category'] = df ['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins = 20, edgecolor ='Black',color= 'red')
plt.xticks(rotation=90)
plt.show()

# below grpah you can see the most of the buyers are T-shirt
```

```python
#checking B2b data by using Pie chart
B2B_Check= df['B2B'].value_counts()

# plot the pie charts
plt.pie(B2B_Check, labels=B2B_Check.index, autopct ='%1.1f%%')
#plt.axis('equal')
plt.show()

#from below chart we can see the maximum 99.3% of buyers are retailers and 0.7% are B2B buyers
```

```python
#Scatter Plot
X_data =df['Category']
Y_data =df['Size']

#plot the Scatter plot
plt.scatter(X_data,Y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter plot')
plt.show()
```

```python
# Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distributionnof State')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 states
Top_10_state= df['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(Top_10_state.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distributionnof State')
plt.xticks(rotation=45)
plt.show()

#note: from below graph you can see that most of the buyers are Maharashtra state
```

```python
#Conclusion
#The data analysis reveals that the business has a significant customer base in the Maharashtra state, primarily serving retailers. 
#It fulfills orders through Amazon, indicating a high demand for t-shirts, with medium size being the preferred choice among buyers.
```
